```python
lines = []
max_size  = 6
with open('/content/raw_30s.tsv', 'r') as fid:
    for i in range(1000):
        line = fid.readline()
        if line == '':
            break
        line = line.split('\t')
        if len(line)<max_size:
                line += [''] * (max_size - len(line))
        lines.append(line)
for i in range(20):
    lines[0].append('a' + str(i))

data = pd.DataFrame(lines[1:], columns=lines[0])

def dealer(X):
    return X[6:]
def dealer2(X):
    return X[7:]
def inttostr(x):
    return str(x)
data['ARTIST_ID'] = data['ARTIST_ID'].apply(lambda x: return x [6:])
data['TRACK_ID'] = data['TRACK_ID'].apply(dealer)
data['ALBUM_ID'] = data['ALBUM_ID'].apply(dealer2)
data['ARTIST_ID'] = data['ARTIST_ID'].apply(inttostr)
data['TRACK_ID'] = data['TRACK_ID'].apply(inttostr)
data['ALBUM_ID'] = data['ALBUM_ID'].apply(inttostr)
data.head(10)
```
